```python

class LinearRegressionModel():
    def __init__(self, learning_rate, epoch):
        self.learning_rate = learning_rate
        self.epoch = epoch
        self.cost = []
        
    def Fit(self, X_train, y_train):
        self.X_train = np.copy(X_train)
        self.y_train = np.copy(y_train)
    
    def GradientDescent(self, X_train, y_train):
        weight = np.random.randn(X_train.shape[1], 1)
        bias = 1
        num_of_DataPoint = len(X_train)

        for epo in range(self.epoch):
            
            y_predicted = X_train.dot(weight) + bias
            y_train = y_train.reshape(y_train.shape[0],1)
            
            error = y_predicted - y_train
            cost_function = (1 / (2 * num_of_DataPoint)) * np.sum(error ** 2)
#             cost_function = 1 - (np.sum(error**2) / (sum((y_train - y_train.mean())**2)))
            self.cost.append(cost_function)
            
            
            weight_derivative = (1 / num_of_DataPoint) * ((X_train.T).dot((error)))
            weight = weight - (self.learning_rate * weight_derivative)
            
            bias_derivative = (1 / num_of_DataPoint) * sum(error)
            bias = bias - (self.learning_rate * bias_derivative)
                
            self.weight = weight
            self.bias = bias
            
            print('iteration: \n{}'.format(epo))
            print()
            print('weight: \n{}'.format(weight))
            print()
            print('bias: \n{}'.format(bias))
            print()
            print('cost: \n{}'.format(cost_function))    
            print('-------------------------------------\n') 
    
    def Predict(self, X_test):
        LinearRegressionModel.GradientDescent(self, self.X_train, self.y_train)
        prediction = X_test.dot(self.weight) + self.bias
        return prediction
```
