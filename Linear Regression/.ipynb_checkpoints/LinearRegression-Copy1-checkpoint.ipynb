```python
class LinearRegressionModel():
    def __init__(self, learning_rate, epoch):
        self.learning_rate = learning_rate
        self.epoch = epoch
        
    def Fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
    
    def GradientDescent(self, X_train, y_train):
        weight = np.ones((X_train.shape[1], 1))
        bias = 1
        num_of_DataPoint = len(X_train)
        
#         print('X_train: \n{}\n'.format(X_train))
#         print('X_train''shape: {}\n'.format(X_train.shape))
#         print('weight: \n{}\n'.format(weight))
        
#         y_predicted = X_train.dot(weight) + bias
#         print('y_predicted: \n{}\n'.format(y_predicted))
#         print('y_predicted''s shape: {}\n'.format(y_predicted.shape))
        
#         print('y_train: \n{}\n'.format(y_train))
#         print('y_train shape: {}\n'.format(y_train.shape))
        
        
#         y_predicted = y_predicted.T
#         print(('error shape: {}\n'.format((y_train - y_predicted).shape)))
        
        
        
        
        for epo in range(self.epoch):
            y_predicted = X_train.dot(weight) + bias
            y_predicted = y_predicted.T
#             y_predicted_transpose = y_predicted.transpose()
#             y_predicted_result = sum(y_predicted_transpose)
            
            
            
#             print("This is y_predicted_result: \n{}".format(y_predicted_result))
#             print()
#             print('this is y_train: \n{}'.format(y_train))
#             print()
#             print('this is y_train - y_predicted_resul: \n{}'.format(y_train - y_predicted_result))
            cost_function = (1 / num_of_DataPoint) * sum(value**2 for value in (y_train - y_predicted))
            
            for i in range(X_train.shape[1]):
                weight_derivative = -(2/ num_of_DataPoint) * ((y_train - y_predicted).dot(X_train[:,i])) #we must ith feture of n data points to compute derivative 
                weight[i,0] = weight[i,0] - (self.learning_rate * weight_derivative)
                
                bias_derivative = -(2/ num_of_DataPoint) * sum(y_train - y_predicted)
                bias = bias - (self.learning_rate * bias_derivative)
                
            self.weight = weight
            self.bias = bias
            
#             print('iteration: \n{}'.format(epo))
#             print()
#             print('weight: \n{}'.format(weight))
#             print()
#             print('bias: \n{}'.format(bias))
#             print()
#             print('cost: \n{}'.format(cost_function))    
#             print('-------------------------------------\n') 
    
    def Predict(self, X_test):
        LinearRegressionModel.GradientDescent(self, self.X_train, self.y_train)
        prediction = sum((X_test.dot(self.weight) + self.bias).transpose())
        return prediction
```
